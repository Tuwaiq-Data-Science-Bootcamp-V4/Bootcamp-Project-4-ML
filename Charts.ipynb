```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import LabelEncoder
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
#pie = px.pie(df, names="Class", color_discrete_sequence=['#DDD2BB','green'])
# pie = px.pie(df, names="Class", color_discrete_sequence=['#357C3C','#BB2525'])
pie = px.pie(df, names="Class", color_discrete_sequence=['forestgreen','red'])
pie.update_traces(textinfo='percent+label')
pie.update_layout(title_text='Edible vs Poisonous', font=dict(size=13))
pie.show()
```

```python
his = px.histogram(df, x='bruises',color='Class',barmode='group', color_discrete_sequence=['red','forestgreen'],
labels={'bruises':'Bruises'}, title='Distribution of Mushrooms by their Classes and Bruises')
his.update_layout(yaxis_title='Number of Mushrooms', font=dict(size=13))
his.show()
```

```python
plt.figure(figsize=(10,5))
fig = sns.countplot(df, x='cap_shape', hue='Class', order=df['cap_shape'].value_counts().index,
palette=['red','forestgreen'], linewidth=2, )

fig.set_xticklabels(['Convex','Flat','Knobbed','Bell','Sunken', 'Conical'])

plt.legend(['Poisonous','Edible'], loc='upper right')
plt.xlabel('Cap Shapes', fontsize=12)
plt.ylabel('Number of Mushrooms', fontsize=12)
plt.title('Distribution of Mushrooms by their Classes and Cap Shapes', fontsize=15)
plt.show()
```

```python
fig = px.histogram(df, x='cap_surface',color='Class', barmode='group', labels={'cap_surface':'Cap Surfaces'},
 title='Distribution of Mushrooms by their Classes and Cap Surfaces', color_discrete_sequence=['red','forestgreen'])
fig.update_layout(font=dict(size=13), yaxis_title='Number of Mushrooms')
fig.show()
```

```python
fig = px.histogram(df, y='cap_color',color='Class', orientation='h', labels={'cap_color':'Cap Colors'},
title='Distribution of Mushrooms by their Classes and Cap Colors', color_discrete_sequence=['red','forestgreen'])
fig.update_layout(font=dict(size=13), xaxis_title='Number of Mushrooms')
fig.show()
```

```python
fig = px.histogram(df, x='habitat', color='Class', pattern_shape='Class', title='Habitat',
 color_discrete_sequence=['red','forestgreen'], labels={'habitat':'Habitat'})

fig.update_layout(font=dict(size=13), yaxis_title='Number of Mushrooms', title='Distribution of Mushrooms by their Classes and Habitats')
fig.show()
```
